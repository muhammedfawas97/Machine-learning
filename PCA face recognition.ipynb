```python
import numpy as np
from os import walk,path
import mahotas as mh
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
x=[]
y=[]
for dir_path,dir_name,file_names in walk('E:\\ML\\att_faces'):
    print(dir_path)
    print(dir_name)
    print(file_names)
```

```python
for dir_path,dir_name,file_names in walk('E:\\ML\\att_faces'):
    for fn in file_names:
        if fn[-3:]=='pgm':
            image_file_name=path.join(dir_path,fn)
            x.append(scale(mh.imread(image_file_name,as_grey=True).reshape(10304).astype('float')))
            y.append(dir_path)

            
X=np.array(x)
print(X.shape)
print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
pca = PCA(n_components=200)

X_train_reduced = pca.fit_transform(X_train)
X_test_reduced = pca.transform(X_test)
print ('The original dimensions of the training data were', X_train.shape)
print ('The reduced dimensions of the training data are', X_train_reduced.shape)
classifier = LogisticRegression()
classifier.fit(X_train_reduced, y_train)
accuracies = cross_val_score(classifier, X_train_reduced, y_train)
print ('Cross validation accuracy:', np.mean(accuracies), accuracies)
predictions = classifier.predict(X_test_reduced)
print (classification_report(y_test, predictions))

```
