```python
df1['Total_income']=df1['ApplicantIncome']+df1['CoapplicantIncome']
cat_col=['Gender','Dependents', 'Married','Education','Self_Employed','Property_Area', 'Loan_Status']
num_col=['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term','Total_income','Credit_History']
for i in cat_col:
    df1[i]=df1[i].fillna(df1[i].mode()[0])
    
for j in num_col:
    df1[j]=df1[j].fillna(df1[j].mean())

#print(df1.isnull().sum())   

df1['Dependents']=df1['Dependents'].apply(lambda x:x.split('+')[0])
#df1['Dependents'].unique()
```

```python
cols=['Gender', 'Married','Education','Self_Employed','Property_Area', 'Loan_Status']
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for col in cols:
    df1[col]=le.fit_transform(df1[col])
    
```

```python
df2=df1.drop(['Loan_ID','ApplicantIncome','CoapplicantIncome'],axis=1)
#df2
```

```python
x=df2.drop(['Loan_Status'],axis=1)
y=df2['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.model_selection import cross_val_score
def classify(model,x,y):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
    model.fit(x_train,y_train)
    print('Accuracy is:',model.score(x_test,y_test))
    score=cross_val_score(model,x,y,cv=5)
    print('CRval:',np.mean(score))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble.forest import RandomForestClassifier,ExtraTreesClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble.forest import RandomForestClassifier,ExtraTreesClassifier


models=[DecisionTreeClassifier(max_depth=5),LogisticRegression(),RandomForestClassifier(n_estimators=100),ExtraTreesClassifier()]

model_name=['DecisionTree','LogisticRegression','Randomforest','Extratree']

for i,model in enumerate(models):
    print()
    print(model_name[i])
    print()
    classify(model,x,y)
    
```
