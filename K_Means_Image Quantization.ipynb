```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import mahotas as mh

img=mh.imread('C:\\Users\\Dell\\Downloads\\Garden.JPEG')/255 ##normalisation
print(img)
#print(img.shape)
#print(img[0])
#print(len(img[0]))
```

```python
from sklearn.utils import shuffle
print(img.ndim)
width,height,depth=tuple(img.shape)
image_flattened=np.reshape(img,(width*height,depth))
print(image_flattened)
print(image_flattened.shape)
```

```python
image_array_sample=shuffle(image_flattened,random_state=0)[:1000]
print(image_array_sample.shape)
```

```python
estimator=KMeans(n_clusters=64,random_state=0)
estimator.fit(image_array_sample)
```

```python
print(estimator.labels_)
```

```python
cluster_assingments=estimator.predict(image_flattened)
print(cluster_assingments)
print(len(cluster_assingments))
compressed_palette=estimator.cluster_centers_
print(len(compressed_palette))
print(compressed_palette)
```

```python
compressed_img=np.zeros((width,height,compressed_palette.shape[1]))
#print(compressed_img)
print(compressed_img.shape)
```

```python
label_idx=0
for i in range(width):
    for j in range(height):
        compressed_img[i][j]=compressed_palette[cluster_assingments[label_idx]]
        label_idx+=1
```

```python
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Compressed Image')
plt.imshow(compressed_img)
plt.axis('off')
plt.show()
plt.imsave('compressedimg.jpg',compressed_img)
```
