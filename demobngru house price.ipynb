```python
df3=df2.dropna()
df3.isnull().sum()
```

```python
df3['BHK']=df3['size'].apply(lambda x:x.split(' ')[0])
df3
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True


```

```python
df3[~df3['total_sqft'].apply(is_float)]
```

```python
def conrt_total_sqft(x):
    Tokens=x.split('-')
    if len(Tokens) == 2:
        return (float(Tokens[0])+float(Tokens[1]))/2
    try:
        return float(x)
    except:
        return None
    
```

```python
df4=df3.copy()
df4['total_sqft']=df3['total_sqft'].apply(conrt_total_sqft)
df4.head(10)
```

```python
df5=df4.copy()
df5['price_per_sqft']=df4['price']*100000/df4['total_sqft']
df5
```

```python
df5.location=df5.location.apply(lambda x:x.strip(' '))
```

```python
location_stats=df5.groupby('location')['location'].agg('count').sort_values(ascending=False)
len(location_stats[location_stats<=10])
```

```python
location_less_than_10=location_stats[location_stats<=10]
location_less_than_10
```

```python
df5.location=df5.location.apply(lambda x:'others' if x in location_less_than_10 else x)
len(df5.location.unique())
```
