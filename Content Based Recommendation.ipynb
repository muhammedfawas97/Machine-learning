```python
import pandas as pd
films=pd.read_csv('E:\\ML\\FilmCB.txt',index_col=1)
films
```

```python
users=pd.read_csv('E:\\ML\\UserCB.txt',index_col=1)
users
```

```python
import numpy as np,sys

def content_recomm(user,user_df,film_df):
    u=user_df.iloc[user][1:].values
    u_norm=np.linalg.norm(u)
    film_recomm=[]
    for row in range(film_df.shape[0]):
        f=film_df.iloc[row][1:].values
        film_name=film_df.index[row]
        f_norm=np.linalg.norm(f)
        s=np.dot(u,f)/(u_norm*f_norm)
        if s>0.8:
            film_recomm.append((film_name,s))
            
    print('Recommendations :',film_recomm)
    film_recomm=sorted(film_recomm,key=lambda x:x[1],reverse=True)
    return film_recomm

r1=content_recomm(0,users,films)
print("Sorted Recommendation :",r1)

        
```
