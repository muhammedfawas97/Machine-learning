```python
import pandas as pd

data=pd.read_csv('C:\\Users\\Dell\\Downloads\\winequality-red.csv',sep=';')
data
```

```python
print(data.shape)
print(data.columns)
print(data.quality.unique())
```

```python
data.isnull().sum()
```

```python
x=data.iloc[:,:-1]
x
```

```python
y=data.iloc[:,-1].values
print(y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
x_scaled=StandardScaler()
x_new=x_scaled.fit_transform(x)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_new,y,test_size=0.2)
Clf=DecisionTreeClassifier()
model=Clf.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
for i ,pred in enumerate(y_pred):
    print('Predicted : %s , actual : %s'%(pred,y_test[i]))
```

```python
print(y_test.shape)
print(y_pred.shape)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
clf2=RandomForestClassifier(n_estimators=100)
model2=clf2.fit(x_train,y_train)
pred2=model2.predict(x_test)
print(classification_report(y_test,pred2))
```

```python
from sklearn.ensemble.forest import ExtraTreesClassifier
from sklearn.ensemble.weight_boosting import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier

models=[ExtraTreesClassifier(n_estimators=100),AdaBoostClassifier(n_estimators=100),BaggingClassifier(n_estimators=100)]
for i,model in enumerate(models):
    Clf3=model.fit(x_train,y_train)
    predicted=Clf3.predict(x_test)
    print(classification_report(y_test,predicted))
```

```python
import sklearn.model_selection as ms
from sklearn import neighbors
from sklearn.model_selection import GridSearchCV
k_neighbours=list(range(1,21,2))
n_grid=[{'n_neighbors':k_neighbours}]
model=neighbors.KNeighborsClassifier()
cv_knn=GridSearchCV(estimator=model,param_grid=n_grid,
                   cv=ms.KFold(n_splits=10))
cv_knn.fit(x2T,y2T)
```

```python
best_k=cv_knn.best_params_['n_neighbors']
print(best_k)
knnclf=neighbors.KNeighborsClassifier(n_neighbors=best_k)
knnclf.fit(x2T,y2T)
yp=knnclf.predict(xTT)
print(classification_report(yp,yTT))
```
