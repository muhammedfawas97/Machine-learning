```python
import pandas as pd
import numpy as np

A=pd.read_csv('E:\\ML\\utility.csv',index_col=0,na_values=['?'])
A
```

```python
print(A.mean())
A.fillna(A.mean(),inplace=True)
A
```

```python
import sys
def cf_recomm(user,book,dat,S=2):
    try:
        u_ind=dat.index.get_loc(user)
        b_ind=dat.columns.get_loc(book)
        #print(u_ind)
        #print(b_ind)
    except:
        print('Error : User/Item doesnt exist')
    u,s,v=np.linalg.svd(dat)
    #print(u)
    #print(v)
    #print(s)
    reduced_sig=np.mat(np.eye(S)*s[:S])
    #print(reduced_sig)
    reduced_u=u[:,:S]
    redused_v=v[:S,:]
    recom=reduced_u[u_ind,:]*reduced_sig*np.mat(redused_v[:,b_ind]).T
    #print(recom)
    return recom.item(0)
bob_score=cf_recomm('Bob','The Martian',A)    
Eve_score=cf_recomm('Eve','Solaris',A)  
print(bob_score)
print(Eve_score)
```
