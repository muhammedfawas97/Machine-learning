```python
import pandas as pd
df=pd.read_csv('C:\\Users\\Dell\\Downloads\\breast-cancer-wisconsin (2).csv',header=None)
#df
```

```python
dfnew=df[df.values=='?']
dfnew
```

```python
x=df.drop([6,10],axis=1)
y=df[10]
```

```python
from sklearn.preprocessing import LabelEncoder
LB=LabelEncoder()
y=LB.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=20)
```

```python
from sklearn.linear_model.logistic import LogisticRegression
Classifier=LogisticRegression()
Classifier.fit(x_train,y_train)
pred=Classifier.predict(x_test)
pred
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as np
pen_val=['l1','l2']
c_val=2.**np.arange(-5,10,step=2)
import sklearn.model_selection as ms
grid=[{'penalty':pen_val,
      'C':c_val}]
model=LogisticRegression(solver='liblinear')
gridcv=GridSearchCV(estimator=model,param_grid=grid,
                   n_jobs=-1,verbose=-1,cv=ms.KFold(n_splits=10))
gridcv.fit(x_train,y_train)

best_c=gridcv.best_params_['C']
best_penalty=gridcv.best_params_['penalty']
print(best_c)
print(best_penalty)
```

```python
b_clf=LogisticRegression(C=best_c,penalty=best_penalty,solver='liblinear')
b_clf.fit(x_train,y_train)
predict=b_clf.predict(x_test)
y_proba=b_clf.predict_proba(x_test)
#print(y_test)
print(predict)
print(y_proba)
print(b_clf.score(x_test,y_test))
print(b_clf.coef_)
print(np.exp(b_clf.coef_))
```

```python
#%pylab
import matplotlib.pyplot as plt

from sklearn.metrics import roc_curve,auc
fpr,tpr,threshold=roc_curve(y_test,y_proba[:,1])
plt.plot(fpr,tpr)
plt.show()
print(auc(fpr,tpr))
```
